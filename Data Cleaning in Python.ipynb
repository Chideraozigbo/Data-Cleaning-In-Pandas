```python
# importing pandas librabry
import pandas as pd
```

```python
# To read the data
data = pd.read_excel("Customer Call List.xlsx")
data
```

```python
# To count the total number of records
data.shape[0]
```

```python
# To drop duplicate records
data = data.drop_duplicates()
```

```python
# To count the total number of records after dropping
data.shape[0]
```

```python
# To drop the "Not_Useful_Column"
data = data.drop(columns = "Not_Useful_Column")
data
```

```python
# To remove /,...,_ from the last_name column
data["Last_Name"] = data["Last_Name"].str.strip("/")
data["Last_Name"] = data["Last_Name"].str.strip("...")
data["Last_Name"] = data["Last_Name"].str.strip("_")
data
```

```python
# To format the Phone_Number column
data['Phone_Number'] = data['Phone_Number'].str.replace('-', '')
data['Phone_Number'] = data['Phone_Number'].str.replace('/', '')
data['Phone_Number'] = data['Phone_Number'].str.replace('|', '')
data
```

```python
# To replace all record in the phone number column with Na to blank
data['Phone_Number'] = data['Phone_Number'].str.replace('Na', '')
data
```

```python
# To split the Address column
data[['Street', 'City', 'Zip_Code']] = data['Address'].str.split(',', n=2, expand=True)
data
```

```python
# To drop the Address column
data = data.drop(columns = "Address")
data
```

```python
# To format the Paying Customer column
data["Paying Customer"] = data["Paying Customer"].str.replace('Yes', 'Y')
data["Paying Customer"] = data["Paying Customer"].str.replace('No', 'N')
data["Paying Customer"] = data["Paying Customer"].str.replace('N/a', '')
data
```

```python
# To format the Do Not Contact column
data["Do_Not_Contact"] = data["Do_Not_Contact"].str.replace('Yes', 'Y')
data["Do_Not_Contact"] = data["Do_Not_Contact"].str.replace('No', 'N')
data
```

```python
# To fill all null records to blank
data.fillna('')
```

```python
# To save the cleaned data

data.to_excel("Transformed_Customer_Call_List.xlsx", index=False)
```
